# 4. 데이터 가공

### 4-1. 생활물가지수

```python
import pandas as pd
import matplotlib.pylab as plt
```

```python
A = pd.read_csv("cost of living index.csv")
AA = A.iloc[0:7, 1:8]
AA = AA.transpose()
AA.columns = ["total index", "cost of living index ","food", "foodX", "monthly", "monthlyX", "living index"]
AAA = AA.drop(['품목별'])
AAA
```

```python
AAA.plot.scatter("total index","food")
AAA.plot.scatter("monthly","living index")
```

### 4-2: 건설공사비 지수

```python
B = pd.read_csv("Construction cost index.csv")
BB = B.iloc[3:26:9, 3:8]
BB.columns = ["January22", "November22", "December22", "January23", "february23"]
BBB = BB.transpose()
BBB.columns = ["residential", "port facility ","power facility"]
BBB
```

```python
BBB.plot.scatter("residential","power facility")
```

## 5. 분석결과


### 5-1. 가설1에 대한 분석결과

- 가설1.생활물가지수는 꾸준하게 증가했을 것이다.
- 설명: 근소한 차이지만 생활물가지수가 올랐다는 사실을 알 수 있다. 

### 5-2. 가설2에 대한 분석결과

- 가설2.건설공사비지수는 꾸준하게 증가했을 것이다.
- 설명: 가설1과 마찬가지로 근소한 차이지만 건설공사비지수가 증가했다는 사실을 알 수 있다. 

## 6. 결론

- 생활물가지수와 건설공사비지수 모두 근소하지만 꾸준하게 증가했다는 사실을 알 수 있다. 즉 가설1과 가설2는 참이다.
